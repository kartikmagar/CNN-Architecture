## (1) What is the role of filters and feature maps in Convolutional Neural Network (CNN)?

Ans :
### 1. Filters (also called kernels)
- A filter is a small matrix of learnable weights (e.g., 3√ó3, 5√ó5).

- It slides (convolves) over the input image (or feature map from the previous layer).

- At each step, it performs a dot product between the filter weights and the patch of input pixels, producing a number.

Purpose :
- Detects local patterns such as edges, corners, textures, or shapes.

- Each filter specializes in a different kind of feature (e.g., vertical edges, horizontal edges).

### 2. Feature Maps (also called activation maps)
- A feature map is the output produced when a filter is applied to the input.

- Each feature map highlights where in the input image a particular feature (edge, shape, etc.) was detected.

- After convolution, the values in the feature map go through:

- Non-linear activation function (e.g., ReLU) ‚Üí adds non-linearity

- Pooling (optional) ‚Üí reduces dimensions while keeping important info

### 3. Combined Role
- Filters extract features (what to look for).

- Feature maps show where those features are found in the input.

- As you stack more convolutional layers:

- Lower layers ‚Üí detect simple features (edges, textures).

- Higher layers ‚Üí detect complex features (faces, objects, shapes).

## (2) Explain the concepts of padding and stride in CNNs(Convolutional Neural Network). How do they affect the output dimensions of feature maps?

Ans:

### 1. Padding
Definition:
Padding means adding extra rows/columns (usually filled with zeros) around the input image before applying the filter.

Why use padding?
- Preserve spatial dimensions ‚Äì Without padding, feature maps shrink after each convolution. (An input of 5√ó5 with a 3√ó3 filter gives a 3√ó3 output).
- Capture edge information ‚Äì Filters can slide even on border pixels.
- Control output size ‚Äì Padding allows you to decide how much reduction happens.

Types of padding:
-
Valid padding ("no padding") ‚Üí shrink output.
Formula:
O = S(W‚àíF)‚Äã+1
- Same padding ("zero padding") ‚Üí keeps output size same as input.
Formula:
O = S/W‚Äã

### 2. Stride
Definition:

- Stride is how many steps the filter moves at a time when sliding over the input.

Effect:

- Stride = 1 ‚Üí filter moves one pixel at a time ‚Üí dense feature map, larger output.
- Stride = 2 ‚Üí filter jumps two pixels ‚Üí smaller feature map, reduces computation.

### 3. Effect on Output Dimension
For an input of size W √ó W, filter size F, padding P, and stride S:
O = S/(W‚àíF+2P)‚Äã+1
Where:

- O = output size (width/height of feature map)

- W = input size

- F = filter size

- P = padding

- S = stride

## (3) Define receptive field in the context of CNNs. Why is it important for deep architectures?

Ans :
### 1. Definition
- The receptive field of a neuron in a CNN is the region of the input image that affects (or contributes to) the neuron‚Äôs activation.

In other words:

- Shallow layers ‚Üí each neuron ‚Äúsees‚Äù only a small patch of the input (like edges or corners).

- Deeper layers ‚Üí neurons combine information from many previous layers, so their receptive field covers a larger portion of the original image (like eyes, faces, or whole objects).

### 2. Example
Suppose:

Input image = 32√ó32

Conv1: filter = 3√ó3
3√ó3, stride = 1 ‚Üí each neuron in layer 1 sees 3√ó3 region.

Conv2: filter = 3√ó3
3√ó3, stride = 1 ‚Üí each neuron in layer 2 sees a 5√ó5 region of original input (since Conv2 neurons depend on Conv1 outputs, which already look at 3√ó3).

Conv3: filter = 3√ó3
3√ó3, stride = 1 ‚Üí receptive field becomes 7√ó7.

## (4) Discuss how filter size and stride influence the number of parameters in a CNN.

Ans :
### 1. Parameters in a Convolutional Layer
- Each convolutional filter (kernel) has its own weights.
- For a filter of size ùêπ√óùêπ applied on input with ùê∂
- C channels (e.g., RGB = 3):

Parameters¬†per¬†filter=F√óF√óC+1¬†(bias)

- If there are K filters:
Total¬†parameters=K√ó(F√óF√óC+1)

### 2. Role of Filter Size
- Larger filter (e.g., 7√ó7 vs 3√ó3) ‚Üí more weights per filter ‚Üí more parameters.
- Example (input depth = 3, one filter):
3√ó3√ó3+1=28 parameters.
7√ó7√ó3+1=148 parameters.
- Larger filters ‚Üí more capacity, but heavier and prone to overfitting.
- Smaller filters (e.g., 3√ó3 stacked multiple times) ‚Üí fewer parameters, but still capture large receptive fields when stacked.

### 3. Role of Stride
- Stride controls how the filter moves ‚Üí affects output feature map size, but does not change number of parameters.
Example:

- Same filter (3√ó3√ó3) with stride 1 or stride 2 ‚Üí still 28 parameters.

- Difference: stride 1 ‚Üí larger feature map, stride 2 ‚Üí smaller feature map (less computation, but same weights).


## (5) Compare and contrast different CNN-based architectures like LeNet,AlexNet, and VGG in terms of depth, filter sizes, and performance.

Ans :
### 1. LeNet (1998, Yann LeCun)
- Purpose: Handwritten digit recognition (MNIST).

- Depth: ~5‚Äì7 layers (2 conv + 2 subsampling + 2 FC).

Filter Sizes:

Conv1 ‚Üí 5√ó5 filters

Conv2 ‚Üí 5√ó5 filters

Key Features:

- Used tanh/sigmoid activations.

- Average pooling (subsampling).

Performance:

- Great for small grayscale images (28√ó28).

- Not scalable to large, colored images (like ImageNet).

### 2. AlexNet (2012, Alex Krizhevsky)
- Purpose: ImageNet Large Scale Visual Recognition Challenge (ILSVRC).

- Depth: 8 layers (5 conv + 3 FC).
Filter Sizes:
- Conv1 ‚Üí 11√ó11 filters, stride 4 (to capture big features quickly).

- Later layers ‚Üí 5√ó5, then multiple 3√ó3 filters.
Key Features:
- Introduced ReLU activation (much faster training than sigmoid/tanh).

- Used Dropout in FC layers to prevent overfitting.

- First to use GPU training effectively.
Performance:
- Won ILSVRC 2012 with top-5 error ~15% (huge improvement over 26%).

- Much deeper & larger than LeNet.

### 3. VGGNet (2014, Oxford Visual Geometry Group)
- Purpose: ImageNet classification (ILSVRC).

- Depth: 16‚Äì19 layers (VGG-16, VGG-19).
Filter Sizes:
- All conv layers use 3√ó3 filters (very small, consistent).

- Stacked multiple 3√ó3 convolutions ‚Üí larger receptive field (e.g., 2 √ó 3√ó3 ‚âà 5√ó5).
Key Features:
- Very deep, very uniform architecture (only 3√ó3 conv + 2√ó2 max pooling).

- Uses ReLU activation.
Performance:
- Achieved top-5 error ~7.3% on ImageNet.

- Computationally heavy (138 million parameters in VGG-16).
- Formed the basis for modern architectures (ResNet, Inception, etc.).

```python
# 6
# Import libraries
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

# 1. Load and preprocess MNIST dataset
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

# Reshape to (samples, 28, 28, 1) since CNN expects 4D input
x_train = x_train.reshape(-1, 28, 28, 1).astype("float32") / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype("float32") / 255.0

# Convert labels to one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# 2. Build CNN model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation="relu"),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dense(10, activation="softmax")
])

# 3. Compile the model
model.compile(optimizer="adam",
              loss="categorical_crossentropy",
              metrics=["accuracy"])

# 4. Train the model
history = model.fit(x_train, y_train, epochs=5, batch_size=64,
                    validation_data=(x_test, y_test))

# 5. Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\nTest accuracy: {test_acc:.4f}")
```

```python
# 7

# Import libraries
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# 1. Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()

# Normalize pixel values (0-255 ‚Üí 0-1)
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Convert labels to one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

print("Training data shape:", x_train.shape)
print("Test data shape:", x_test.shape)

# 2. Build CNN model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation="relu", padding="same", input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation="relu", padding="same"),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128, (3,3), activation="relu", padding="same"),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.5),  # prevent overfitting
    layers.Dense(10, activation="softmax")
])

# 3. Compile the model
model.compile(optimizer="adam",
              loss="categorical_crossentropy",
              metrics=["accuracy"])

# 4. Train the model
history = model.fit(x_train, y_train, epochs=10, batch_size=64,
                    validation_data=(x_test, y_test))

# 5. Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\nTest accuracy: {test_acc:.4f}")

# 6. Plot training history
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
# 8

# Import libraries
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# 1. Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 2. Data preprocessing & loading
transform = transforms.Compose([
    transforms.ToTensor(),  # Convert to tensor
    transforms.Normalize((0.1307,), (0.3081,))  # Normalize MNIST
])

train_dataset = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root="./data", train=False, download=True, transform=transform)

train_loader = DataLoader(dataset=train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=1000, shuffle=False)

# 3. Define CNN model
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)   # (1 ‚Üí 32 filters)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)  # (32 ‚Üí 64 filters)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(64 * 7 * 7, 128)  # after 2x pooling, 28x28 ‚Üí 7x7
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))  # Conv1 + ReLU + Pool
        x = self.pool(F.relu(self.conv2(x)))  # Conv2 + ReLU + Pool
        x = x.view(-1, 64 * 7 * 7)            # Flatten
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = CNN().to(device)

# 4. Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 5. Training loop
epochs = 5
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()        # Reset gradients
        outputs = model(images)      # Forward pass
        loss = criterion(outputs, labels)
        loss.backward()              # Backpropagation
        optimizer.step()             # Update weights

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(train_loader):.4f}")

# 6. Evaluation
model.eval()
correct, total = 0, 0
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"\nTest Accuracy: {100 * correct / total:.2f}%")
```
